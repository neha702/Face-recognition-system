```
import cv2
```

```
face_cascade=cv2.CascadeClassifier(cv2.data.haarcascades +"haarcascade_frontalface_default.xml") #Cascade classifier(Pre trained)
img=cv2.imread("/content/friends.jpeg") #Reading image
width = int(img.shape[1]/2)  
height = int(img.shape[0]/2)  
dim = (width, height)  
img=cv2.resize(img,dim) #resize image
gray_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #colored image to black and white
```

```
face=face_cascade.detectMultiScale(gray_image,1.05,5) # function to detect images
```

```
from google.colab.patches import cv2_imshow
```

```
#Add rectangular face box
for (x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 3)
# Display the output
cv2_imshow(img)
cv2.waitKey()
```
